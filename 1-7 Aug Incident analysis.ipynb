```python
# function to calculate the indices of duplicate accidents with different messsages
def indices(lst, item):
    return [i for i, x in enumerate(lst) if x == item]

#indices(message_list, "(1/8)9:00")
```

```python
Dupli_Indices = []
L = []
Final_Congested_Accidents = []
for i in range(len(message_list)):
    a = message_list[i]
    Dupli_Indices.append(indices(message_list, a))
for i in range(len(Dupli_Indices)):
    if(Dupli_Indices[i] not in L):
        L.append(Dupli_Indices[i])
for i in range(len(L)):
    if len(L[i])>1:
        Final_Congested_Accidents.append(CongestedAcci[L[i][1]])
      
```
