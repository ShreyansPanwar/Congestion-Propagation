```python

def merge_JsonFiles(filename, month): #function to combine all the data in a day#
    result = list()
    for f1 in filename:
        #can add path to f1 if accessing from other folder
        with open(f1, 'r') as infile:
            result.extend(json.load(infile))

    with open( month + '_combined.json', 'w') as output_file:
        json.dump(result, output_file)

```
